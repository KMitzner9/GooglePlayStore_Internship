## User would like to see how many Entertainment Apps have a Content Rating of Mature

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('clean_googleplaystoremaster.csv')
df.head(10)
```

```python
# take a look at all the categories
df['Category'].unique()
```

```python
# take a look at the content ratings
df['Content Rating'].unique()
```

```python
# should I include adults only in with mature or do they just want to know about the mature category?
```

```python
# make a new dataframe with just the apps in the entertainment category
ent = df.loc[ df['Category'] == 'ENTERTAINMENT' , : ]
ent
```

```python
ent['Content Rating'].unique()
# looks like I dont need to worry about the adult 18+ category
```

```python
ent['Genres'].unique()
```

```python
content_by_category = pd.crosstab(df['Category'], df['Content Rating'])
content_by_category
```

```python
ent_content = pd.DataFrame(content_by_category.loc['ENTERTAINMENT'])
ent_content
```

```python
ent_content = ent_content.iloc[1:5, :]
ent_content
```

```python
ent_content.sum()
```

```python
cat = ent_content['Content Rating']
pos = np.arange(len(cat))
value = ent_content['ENTERTAINMENT']

plt.bar(pos, value)
plt.xticks(pos, cat)
plt.ylabel('Number of Apps')
plt.suptitle('Content Rating of Entertainment Apps')
plt.savefig('ent_content_bar.png')
```

```python
data = [6, 105]

plt.subplots()
plt.suptitle('Content Rating of Entertainment Apps')
plt.pie(data, labels=['Mature 17+ (6)', 'Other (105)'], startangle=90, colors=['#cc6677', '#882255'])
plt.savefig('ent_content_pie.png')
```
